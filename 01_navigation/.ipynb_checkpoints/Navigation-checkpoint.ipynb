````python
import os

import numpy as np
import random
from collections import namedtuple, deque

import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim

import matplotlib.pyplot as plt

device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")

if not os.path.exists("./_checkpoints/"):
    os.mkdir("./_checkpoints/")

if not os.path.exists("./_saved_models/"):
    os.mkdir("./_saved_models/")

def draw_score(scores):
    N = len(scores)
    mv_avg = [sum(scores[i:i+100])/100 for i in range(N-100)]
    
    # plot the scores
    fig = plt.figure(figsize=(15,5))
    plt.plot(np.arange(N), scores, 'b')
    plt.plot(np.arange(100,N),mv_avg, 'r')
    plt.ylabel('Score')
    plt.xlabel('Episode')
    plt.legend(('score', 'moving average 100 scores'))
    plt.show()
````

### 4.1. Define Networks

````python
class QNetwork(nn.Module):
    """Actor (Policy) Model."""

    def __init__(self, state_size, action_size, seed, fc1_units=64, fc2_units=64, fc3_units=64):
        """Initialize parameters and build model.
        Params
        ======
            state_size (int): Dimension of each state
            action_size (int): Dimension of each action
            seed (int): Random seed
            fc1_units (int): Number of nodes in first hidden layer
            fc2_units (int): Number of nodes in second hidden layer
        """
        super(QNetwork, self).__init__()
        self.seed = torch.manual_seed(seed)
        self.fc1 = nn.Linear(state_size, fc1_units).to(device)
        self.fc2 = nn.Linear(fc1_units, fc2_units).to(device)
        self.fc3 = nn.Linear(fc2_units, fc2_units).to(device)
        self.fc4 = nn.Linear(fc2_units, action_size).to(device)
        

    def forward(self, state):
        """Build a network that maps state -> action values."""
        x = F.relu(self.fc1(state))
        x = F.relu(self.fc2(x))   
        x = F.relu(self.fc3(x)) 
        return self.fc4(x)
````

### 4.2. Define Training Procedure

````python
def training(agent, name, n_episodes=2000, max_t=1000, eps_start=1.0, eps_end=0.01, eps_decay=0.995, verbose_every=100):
    """Deep Q-Learning.
    
    Params
    ======
        n_episodes (int): maximum number of training episodes
        max_t (int): maximum number of timesteps per episode
        eps_start (float): starting value of epsilon, for epsilon-greedy action selection
        eps_end (float): minimum value of epsilon
        eps_decay (float): multiplicative factor (per episode) for decreasing epsilon
    """
    scores = []                        # list containing scores from each episode
    scores_window = deque(maxlen=100)  # last 100 scores
    eps = eps_start                    # initialize epsilon
    
    for i_episode in range(1, n_episodes+1):
        beta = min(1, i_episode/1000)
        env_info = env.reset(train_mode=True)[brain_name]
        state = env_info.vector_observations[0]
        score = 0
        for t in range(max_t):
            action = np.int32(agent.act(state, eps))
            env_info = env.step(action)[brain_name]        # send the action to the environment
            next_state = env_info.vector_observations[0]   # get the next state
            reward = env_info.rewards[0]                   # get the reward
            done = env_info.local_done[0]                  # see if episode has finished
            agent.step(state, action, reward, next_state, done)
            score += reward                                # update the score
            state = next_state                             # roll over the state to next time step
            if done:                                       # exit loop if episode finished
                break
                
        scores_window.append(score)       # save most recent score
        scores.append(score)              # save most recent score
        eps = max(eps_end, eps_decay*eps) # decrease epsilon
        agent.episode_end()
        
        print('\rEpisode {}\tMin Score: {:.2f}  \tAverage Score: {:.2f}  \tLast: {:.2f}'.format(i_episode, 
                                                                                                np.min(scores_window), 
                                                                                                np.mean(scores_window), 
                                                                                                scores[-1]), end="")
        if i_episode % verbose_every == 0:
            print('\rEpisode {}\tMin Score: {:.2f}  \tAverage Score: {:.2f}'.format(i_episode, 
                                                                                    np.min(scores_window), 
                                                                                    np.mean(scores_window)))
        
        if i_episode % 1000 == 0:
            torch.save(agent.qnetwork_local.state_dict(), './_checkpoints/checkpoint_{}_{}.pth'.format(name,i_episode))
        
        # Check 100 consecutive scores over 13
        if all(np.array(scores_window) >= 13.0):
            print('\nEnvironment solved in {:d} episodes!\tAverage Score: {:.2f}'.format(i_episode-100, 
                                                                                         np.mean(scores_window)))
            torch.save(agent.qnetwork_local.state_dict(), './_saved_models/checkpoint_{}.pth'.format(name))
            break
    else:
        torch.save(agent.qnetwork_local.state_dict(), './_saved_models/checkpoint_{}.pth'.format(name))
    return scores
````

### 4.5. Parameter Tunning

````python
from Source.ReplayBuffer import ReplayBuffer
from Source.BaseAgent import BaseAgent
import itertools

scores_param = {}
learning_rates = [1e-3, 1e-2]
gammas = [0.99, 0.995]
taus = [1e-3, 5e-3]
params = [(lr,gamma,tau) for lr in learning_rates for gamma in gammas for tau in taus]

for lr, gamma, tau in params:
    print("Learning Rate: {} | Gamma: {} | Tau: {}".format(lr, gamma, tau))
    qnetwork_local = QNetwork(state_size=state_size, action_size=action_size, seed=0)
    qnetwork_target = QNetwork(state_size=state_size, action_size=action_size, seed=1)
    optimizer = optim.Adam(qnetwork_local.parameters(), lr=lr)

    buffer = ReplayBuffer(buffer_size=int(1e5), batch_size=128, seed=2)

    agent = BaseAgent(state_size=state_size, 
                      action_size=action_size, 
                      replay_buffer=buffer,
                      qnetwork_local=qnetwork_local, 
                      qnetwork_target=qnetwork_target, 
                      optimizer=optimizer,
                      gamma=gamma,
                      tau=tau,
                      update_every=4,
                      device=device,
                      seed=3)

    scores_param[(lr, gamma, tau)] = training(agent, "DQN_test", n_episodes=1000)
    print("  -=================================-  \n")
````

````python
# Apparently, lower training rates behave better. 
learning_rates = [1e-4, 5e-4]
gammas = [0.99]
taus = [1e-3]
params = [(lr,gamma,tau) for lr in learning_rates for gamma in gammas for tau in taus]

for lr, gamma, tau in params:
    print("Learning Rate: {} | Gamma: {} | Tau: {}".format(lr, gamma, tau))
    qnetwork_local = QNetwork(state_size=state_size, action_size=action_size, seed=0)
    qnetwork_target = QNetwork(state_size=state_size, action_size=action_size, seed=1)
    optimizer = optim.Adam(qnetwork_local.parameters(), lr=lr)

    buffer = ReplayBuffer(buffer_size=int(1e5), batch_size=128, seed=2)

    agent = BaseAgent(state_size=state_size, 
                      action_size=action_size, 
                      replay_buffer=buffer,
                      qnetwork_local=qnetwork_local, 
                      qnetwork_target=qnetwork_target, 
                      optimizer=optimizer,
                      gamma=gamma,
                      tau=tau,
                      update_every=4,
                      device=device,
                      seed=3)

    scores_param[(lr, gamma, tau)] = training(agent, "DQN_test", n_episodes=1000)
    print("  -=================================-  \n")
````

````python
AGGREGATION_SPAN = 100

legend = []
plt.figure(figsize=(15,5))
for key, scores in scores_param.items():
    score = [sum(scores[i:i+AGGREGATION_SPAN])/AGGREGATION_SPAN for i in range(len(scores) - AGGREGATION_SPAN)]
    plt.plot(np.arange(AGGREGATION_SPAN, len(scores)), score)
    legend.append("LR: {}, Gamma: {}, Tau: {}".format(key[0], key[1], key[2]))
plt.ylabel("Score")
plt.xlabel("Episode")
plt.legend(legend)
plt.show()
````
